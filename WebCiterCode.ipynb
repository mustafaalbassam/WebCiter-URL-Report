```python
#importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importing the dataset
source_df = pd.read_csv('./Sites_URLs_Dataset.csv')
source_df
```

```python
#dropping the additional index column
source_df.drop('index', axis='columns', inplace=True)
source_df
```

```python
#some analysis on the dataset to make sure that there are no missing values

source_df.info()
```

```python
#computing and displaying summary statistics for each feature within the dataframe
source_df.describe()
```

```python
#providing sample of the dataset
source_df.head()
# 1 => Legitimate
# 0 => suspicious
# -1 => fake or phishing
```

```python
# Changing the -1 in "Redirect" to 0, in order to make the 2 classes 1 (Legitimate), and 0 (Fake)

for i in range(0,len(source_df['Result'])):
            
    if source_df['Result'][i] == -1:
            
        source_df["Result"][i] = 0
```

```python
#checking if there is much difference in the counts of the phishing and benign url classes
source_df['Result'].value_counts()
```

```python
# plotting the counts of the Result column

plt.bar([1,0],list(source_df['Result'].value_counts()),color="purple")
plt.title("Results Count")
plt.xlabel("classes")
plt.ylabel("count")
plt.show()
```

```python
# Printing the unique value or "classes" of each column or feature:

for col in source_df.columns:
    unique_list = source_df[col].unique()

    print('{} contains the values:\t\t{}'.format(col,unique_list))
    
# 1 => Legitimate
# 0 => suspicious
# -1 => fake or phishing
```

```python
#Features
X = source_df.iloc[:, : -1].values

#Target 
y = source_df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#splitting the dataset with 20% testing data and 80% training data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 33)

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
#importing the model evaluating functions
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
## Logistic Regression

from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state = 1)

classifier1.fit(X_train, y_train)
```

```python
#Making predictions by using the test data

y_pred1 = classifier1.predict(X_test)
y_pred1
```

```python
#evaluating the model by comparing with the test data

cm1 = confusion_matrix(y_test, y_pred1)
print("Logistic Regression Cofusion matrix:")
print(cm1)
print("\n")
print("\n")
print("Logistic Regression classification report:")
cr1 = classification_report(y_test, y_pred1)
print(cr1)
print("\n")
print("\n")
accuracy1 = accuracy_score(y_test, y_pred1)
print("Logistic Regression accuracy is {:.2%}".format(accuracy1))
```

```python
## K Nearest Neighbors (KNN)

from sklearn.neighbors import KNeighborsClassifier
classifier2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)

classifier2.fit(X_train, y_train)
```

```python
#Making predictions by using the test data
y_pred2 = classifier2.predict(X_test)
y_pred2
```

```python
#evaluating the model by comparing with the test data

cm2 = confusion_matrix(y_test, y_pred2)
print("KNN Cofusion matrix:")
print(cm2)
print("\n")
print("\n")
print("KNN classification report:")
cr2 = classification_report(y_test, y_pred2)
print(cr2)
print("\n")
print("\n")
accuracy2 = accuracy_score(y_test, y_pred2)
print("KNN accuracy is {:.2%}".format(accuracy2))
```

```python
## Support vector machines (SVM)

from sklearn.svm import SVC
classifier3 = SVC(kernel = 'linear', random_state = 0)

classifier3.fit(X_train, y_train)
```

```python
#Making predictions by using the test data

y_pred3 = classifier3.predict(X_test)
y_pred3
```

```python
#Evaluating the model by comparing with the test data

cm3 = confusion_matrix(y_test, y_pred3)
print("SVM Cofusion matrix:")
print(cm3)
print("\n")
print("\n")
print("SVM classification report:")
cr3 = classification_report(y_test, y_pred3)
print(cr3)
print("\n")
print("\n")
accuracy3 = accuracy_score(y_test, y_pred3)
print("SVM accuracy is {:.2%}".format(accuracy3))
```

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier
classifier4 = RandomForestClassifier(n_estimators = 1000, criterion = 'entropy', random_state = 0)

classifier4.fit(X_train, y_train)
```

```python
#Making predictions by using the test data

y_pred4 = classifier4.predict(X_test)
y_pred4
```

```python
#Evaluating the model by comparing with the test data

cm4 = confusion_matrix(y_test, y_pred4)
print("Random Forest Cofusion matrix:")
print(cm4)
print("\n")
print("\n")
print("Random Forest classification report:")
cr4 = classification_report(y_test, y_pred4)
print(cr4)
print("\n")
print("\n")
accuracy4 = accuracy_score(y_test, y_pred4)
print("Random Forest accuracy is {:.2%}".format(accuracy4))
```

```python
#Here, I am comapring the performance of the 4 models using a bar chart

a = ['Logetic\nRegression', 'SVM','KNN', 'Random\nForest']
b = [accuracy1*100, accuracy2*100, accuracy3*100, accuracy4*100]

plt.figure(figsize=(10, 6))
plt.bar(a[0], b[0], color= '#F08080')
plt.bar(a[1], b[1], color= '#9370DB')
plt.bar(a[2], b[2], color= '#4682B4')
plt.bar(a[3], b[3], color= '#20B2AA')
plt.title('Accuracy score of each model')
plt.xlabel('classification model')
plt.ylabel('accuracy %')

plt.show()
```

```python
# Creating a dashboard that let the user checks the URL and see if it is legitimate or phishing


from tkinter import Tk, Button, Label, StringVar, Entry
from PIL import Image,ImageTk # PIL is to serve images
from tkinter import *

window = Tk()
window.title("WebCiter")
window.geometry("800x400")
window.resizable(width=True,height=True)

img_logo = ImageTk.PhotoImage(file='images/Cyber.png')
img=Label(window,image=img_logo).place(x=0,y=0,relwidth=1,relheight=1)

def predict():
    
    ft_value = str(ft_entry.get())
        
    test = []

    ############################################################

    # Using the IP Address

    sp = ft_value.split("//")
    sp = sp[1].split("/")
    sp = sp[0].split(".")

    if len(sp) == 4:
        test.append(-1)

    else:
        test.append(1)
    ############################################################

    # Long URL to Hide the Suspicious Part

    if len(ft_value) < 54:
        test.append(1)

    elif len(ft_value) >= 54 and len(ft_value)<=75:
        test.append(0)

    else:
        test.append(-1)
    ############################################################

    # Using URL Shortening Services “TinyURL”

    if "http" in ft_value:
        test.append(1)

    else:
        test.append(-1)
    ############################################################

    # URL’s having “@” Symbol

    if "@" in ft_value:
        test.append(-1)

    else:
        test.append(1)
    ############################################################

    # Redirecting using “//”

    if ft_value[0:7] == "http://" or ft_value[0:8] == "https://":
        test.append(1)

    else:
        test.append(-1)  
    ############################################################

    # Adding Prefix or Suffix Separated by (-) to the Domain

    if "-" in ft_value.split("/")[2]:
        test.append(-1)

    else:
        test.append(1)      
    ############################################################

    # Sub Domain and Multi Sub Domains

    if len((ft_value.split("/")[2]).split(".")) < 3:
        test.append(1)

    elif len((ft_value.split("/")[2]).split(".")) == 3:
        test.append(0)

    else:
        test.append(-1) 
    ############################################################

    # HTTPS (Hyper Text Transfer Protocol with Secure Sockets Layer)

    test.append(0)
    ############################################################

    # Domain Registration Length

    test.append(0)
    ############################################################

    # Favicon

    test.append(0)
    ############################################################

    # Using Non-Standard Port

    test.append(0)
    ############################################################

    # The Existence of “HTTPS” Token in the Domain Part of the URL

    if "http" in ft_value.split("/")[2] or "https" in ft_value.split("/")[2]:
        test.append(-1)

    else:
        test.append(1)   
    ############################################################

    # Request URL

    test.append(0)
    ############################################################

    # URL of Anchor

    test.append(0)
    ############################################################

    # Links in <Meta>, <Script> and <Link> tags

    test.append(0)
    ############################################################

    # Server Form Handler (SFH)

    test.append(0)
    ############################################################

    # Submitting Information to Email

    test.append(0)
    ############################################################

    # Abnormal URL

    test.append(0)
    ############################################################

    # Website Forwarding

    test.append(0)
    ############################################################

    # Status Bar Customization

    test.append(0)
    ############################################################

    # Disabling Right Click

    test.append(0)
    ############################################################

    # Using Pop-up Window

    test.append(0)
    ############################################################

    # IFrame Redirection

    test.append(0)
    ############################################################

    # Age of Domain

    test.append(0)
    ############################################################

    # DNS Record

    test.append(0)
    ############################################################

    # Website Traffic

    test.append(0)
    ############################################################

    # PageRank

    test.append(0)
    ############################################################

    # Google Index

    test.append(0)
    ############################################################

    # Number of Links Pointing to Page

    test.append(0)
    ############################################################

    # Statistical-Reports Based Feature

    test.append(0)
    ############################################################
    
    if list(classifier4.predict([test]))[0] == 1:
        val = "Legitimate Site"
    else:
        val = "Phishing Site"

    mt_value.set(val)

    
def clear():
    ft_value.set("")
    mt_value.set("")


    
ft_lbl = Label(window,text="Site URL",bg="purple",fg="white",width=14)
ft_lbl.grid(column=0,row=0,padx=15,pady=15)

ft_value = StringVar()
ft_entry = Entry(window,textvariable=ft_value,width=30)
ft_entry.grid(column=1,row=0)
ft_entry.delete(0,'end')

mt_lbl = Label(window,text="URL Analysis",bg="brown",fg="white",width=14)
mt_lbl.grid(column=0,row=1)

mt_value = StringVar()
mt_entry = Entry(window,textvariable=mt_value,width=30)
mt_entry.grid(column=1,row=1,pady=30)
mt_entry.delete(0,'end')



convert_btn = Button(window,text="Show Result",bg="grey",fg='brown',width=14,command=predict)
convert_btn.grid(column=0,row=3,padx=15)
clear_btn = Button(window,text="Clear",bg="black",fg='brown',width=14,command=clear)
clear_btn.grid(column=1,row=3,padx=15)


window.mainloop()
```
